### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Data for all plots
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (top-left)
axs[0, 0].plot(x, y, marker='o', color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Bar chart (top-right)
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'purple'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot (bottom-left)
axs[1, 0].scatter(x, y, color='orange')
axs[1, 0].set_title('Scatter Plot')
# Add annotations
for i in range(len(x)):
    axs[1, 0].annotate(f'({x[i]}, {y[i]})', (x[i], y[i]))

# Pie chart (bottom-right)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Keeps pie chart circular

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
